```python
def prob1(df):
    logisticHubs=df_athens["logistic_hub"].unique()
    arrayLH=[]
    arrayLH2=[]
    arrayBool=[]
    arrayCounter=[]
    arrayProb=[]
    for lh in logisticHubs:
        tmpDf_T=df[(df['logistic_hub']==lh) & (df['late_order']==True)]
        tmpDf_F=df[(df['logistic_hub']==lh) & (df['late_order']==False)]
        arrayLH.extend([lh, lh])
        arrayLH2.append(lh)
        arrayBool.append(True)
        arrayBool.append(False)
        arrayCounter.append(tmpDf_T.shape[0])
        arrayCounter.append(tmpDf_F.shape[0])
        if tmpDf_F.shape[0] == 0:
            arrayProb.append(0)
        else:
            arrayProb.append(tmpDf_T.shape[0]/tmpDf_F.shape[0])
    d2 = {'logistic_hub': arrayLH2,'LateProb': arrayProb}
    df2=pd.DataFrame(data=d2)
    
    return df2
```
